```python
# import library
import os
import pandas as pd
```

```python
base_dir = 'train'
directory = os.fsencode(base_dir)

dfs = []
for file in os.listdir(directory) :
    filename = os.fsdecode(file)
    path = base_dir + '/' + filename

    df = pd.read_csv(path)
    df['sensor'] = int(filename[1:2])
    if (path[:-1] == 'M') :
        df['male'] = 1
    else :
        df['male'] = 0
    
    dfs.append(df)

df_all = pd.concat([df for df in dfs])

df_all.to_csv(base_dir +'.csv')

df_all
```
